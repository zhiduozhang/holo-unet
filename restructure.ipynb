## Code to move heightmaps to single folder for easier testing. Also storing ids for future backtracking

```python
import fnmatch
import scipy.io as sio
import os
import numpy as np
import pandas as pd

from shutil import copyfile
```

```python
PATH = "data\\unlabelled"
id_file_name = 'ids.csv'
```

Load previous database of ids if available

```python
df = None

for file in os.listdir(PATH):
    if fnmatch.fnmatch(file, id_file_name):
        df = pd.read_csv(f'{PATH}\\{file}')

if df is None:
    df = pd.DataFrame(columns=['name', 'experiment', 'indiv_start'])
```

```python
files = []

source_path_2 = "data\\2\\indiv_start\\data"
source_path = "data\\1\\in_sart\\in_sart\\data"

for file in os.listdir(source_path):
    if fnmatch.fnmatch(file, '*.mat'):
        files.append([file, '1', 'indiv_start'])
        copyfile(f'{source_path}\\{file}', f'{PATH}\\{file}')
        
df2 = pd.DataFrame(files, columns=['name', 'experiment', 'indiv_start'])
df = df.append(df2, ignore_index=True)
```
